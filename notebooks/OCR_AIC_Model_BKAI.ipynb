```python
def filter_letter(text: str):
    text_filter = ""
    vowel_letters = "o, ô, ơ, a, â, ă, ê, e, u, ư, i, y,\
    ó, ồ, ớ, á, ấ, ắ, ề, é, ụ, ứ, í, ý, ò, ộ, ờ, à, ầ, ằ, ế, è, ú, ừ, ì, ỳ,\
        õ, ổ, ở, ã, ẫ, ẵ, ễ, ẽ, ũ, ữ, ĩ, ỷ, ỏ, ỗ, ợ, ả, ẩ, ẳ, ể, ẻ, ủ, ử, ỉ, ỹ, ọ, ố, ỡ, ạ, ậ,\
            ặ, ệ, ẹ, ù, ự, ị, ỵ, 1, 2, 3, 4, 5, 6, 7, 8, 9, 0"
    for i in text.split(" "):
        if len(i) <=1 and (i != "y"or i!= "ý" ):
            continue
        elif len(i) == 2:
            num_not_vowel = 0
            for letter in i:
                if letter not in vowel_letters:
                    num_not_vowel +=1
            if num_not_vowel == 2:
                continue # Delete word
            else:
                text_filter += (" "+i)
        elif i == "giay" or i=="giây" or i == "tinchinh" or i == "tinchính" or i == "glay":
            continue
        elif re.search(r'\d{2}:\d{2}', i) is not None:
            continue
        elif "000" in i:
            continue
        else:
            text_filter += (" "+i)
    return text_filter


def filter_text(text: str):
    time_pattern = r'\d{2}:\d{2}'
    match = re.search(time_pattern, text)
    matches = re.findall(time_pattern, text)
    if matches:
        extracted_characters = match.group()+text[text.find(matches[0]) + len(matches[0]) :]
        text_time = extracted_characters
    else:
        text_time = text
    text_return = filter_letter(text_time)
    return text_return
```

```python
path_txt = "/content/info_ocr.txt"

df_txt = pd.read_csv(path_txt, delimiter = ",", header=None)
for idx in range(len(df_txt[2])):
    try:
        text_filter = filter_text(df_txt[2][idx])
        df_txt[2][idx] = str(text_filter)
    except:
        df_txt[2][idx] = "nan"
        print("Current indx:", idx, df_txt[2][idx])

# Xóa hàng trống hoặc ít ký tự (<=5)
num_del = 0
for i in range(len(df_txt[2])):
    if df_txt[2][i]== " " or df_txt[2][i]== "" or len(df_txt[2][i]) <= 5 or df_txt[2][i] == "nan":
        num_del +=1
        df_txt = df_txt.drop(index=i)
print(" Num row delete: ", num_del)
path_save_txt = "/content/info_ocr_processed.txt"
df_txt.to_csv(path_save_txt, sep=',', index=False, header=False)
```

```python
def process_name(name: int):
    return "0"*(6-len(str(name))) + str(name)

def sort_key(file_path):
    file_name = os.path.basename(file_path)
    match = re.match(r'(\D+)(\d+)_(V)(\d+)', file_name)
    if match:
        prefix = match.group(1)
        number = int(match.group(2))
        suffix_number = int(match.group(4))
        return (prefix, number, suffix_number)
    return (file_name, 0, 0)

def write_bin_file_ocr(bin_path: str,npy_path: str ,txt_path: str, method='cosine', feature_shape= 768): # Edit 256, 512, 768
  df_ocr = pd.read_csv(txt_path, delimiter=",", header=None)

  if method in 'L2':
    index = faiss.IndexFlatL2(feature_shape)
  elif method in 'cosine':
    index = faiss.IndexFlatIP(feature_shape)
  else:
    assert f"{method} not supported"

  feats = np.load(npy_path)
  print(len(feats))
  for idx in tqdm.tqdm(range(len(feats))):

 # Output ID 1, 2, 3, 4 . if len path == id reset id
    feat = feats[idx]

    feat = feat.astype(np.float32).reshape(1,-1)
    index.add(feat)


  faiss.write_index(index, os.path.join(bin_path, f"faiss_BKAI_OCR_{method}_hp.bin"))
  print(f"Save {os.path.join(bin_path, f'faiss_BKAI_OCR_{method}_hp.bin')}")
  print(f"Number of Index: {idx}")

write_bin_file_ocr(bin_path = "/content", npy_path = "/content/npy/_ocr.npy", txt_path="/content/info_ocr_processed.txt")
```

```python
import json

# Hàm để tạo file JSON từ file TXT
def txt_to_json(txt_file, json_file):
    keyframes = []

    with open(txt_file, 'r', encoding='utf-8') as file:
        lines = file.readlines()

        for line in lines:
            parts = line.strip().split(',')
            if len(parts) >= 2:
                file_name = parts[0]
                number = parts[1]
                # Tạo đường dẫn cho từng ảnh
                path = f"keyframes/Keyframes_{file_name[:4]}/{file_name}/{number}.jpg"
                keyframes.append(path)

    # Ghi ra file JSON
    with open(json_file, 'w', encoding='utf-8') as json_out:
        json.dump(keyframes, json_out, ensure_ascii=False, indent=4)

# Sử dụng hàm
txt_to_json('/content/info_ocr_processed.txt', '/content/output.json')
```
